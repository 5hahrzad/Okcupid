```python
# Group the data by 'sex' and count the number of occurrences
gender_counts = cupid_df['sex'].value_counts()

# Plot the gender breakdown
sns.barplot(x=gender_counts.values, y=gender_counts.index, palette='viridis')

plt.title('Gender breakdown')
plt.xlabel('Number of users')
plt.ylabel('Gender')

plt.show()
```

```python
# Group the data by 'orientation' and count the number of occurrences
orientation_counts = cupid_df['orientation'].value_counts()

sns.set(style="whitegrid")

sns.barplot(x=orientation_counts.index, y=orientation_counts.values, palette='viridis')

plt.title('Sexual orientation breakdown')
plt.xlabel('Orientation')
plt.ylabel('Number of users')

plt.xticks(rotation=45)

plt.show()
```

```python
# Group the data by 'orientation' and 'sex' and count the number of occurrences
orientation_counts = cupid_df.groupby(['orientation', 'sex']).size().reset_index(name='count')

sns.set(style="whitegrid")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Iterate over the genders and create a bar plot for each gender
for i, gender in enumerate(orientation_counts['sex'].unique()):
    ax = axes[i]  # Select the current axis
    gender_cupid_df = orientation_counts[orientation_counts['sex'] == gender]  # Filter cupid_df for the current gender
    
    # Create the bar plot using seaborn
    sns.barplot(data=gender_cupid_df, x='orientation', y='count', palette='viridis', ax=ax)
    
    # Set the title and axis labels
    ax.set_title(f"Gender: {gender}")
    ax.set_xlabel('Orientation')
    ax.set_ylabel('Number of users')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)  # Rotate x-axis labels for better readability

plt.suptitle('Sexual orientation breakdown by Gender', fontsize=16)

plt.tight_layout()

plt.show()
```

```python
# Filter out users whose age is greater than 90 as they are outliers
data = cupid_df.query("age < 90")

# Dropping all common rows with NA
data_gender_age = cupid_df[["sex", "age"]].dropna()

data_gender_age["sex"] = np.select(
    [
        data_gender_age["sex"] == "m",
        data_gender_age["sex"] == "f"
    ],
    [
        "Male", "Female"
    ],
    default="0")

# Create necessary bins
bins = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
bin_labels = [15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
data_gender_age["bin"] = pd.cut(data_gender_age["age"], bins=bins)

data_gender_age["bin_label"] = data_gender_age["bin"].map(lambda x: str(x.left) + " - " + str(x.right))

data_gender_age["bin"] = data_gender_age["bin"].map(lambda x: x.left)

count_data = data_gender_age.groupby(['sex', 'bin_label']).size().reset_index(name='count')

fig, ax = plt.subplots(figsize=(8, 6))

# Create the pyramid plot using seaborn
sns.barplot(data=count_data, x='count', y='bin_label', hue='sex', palette='muted', ax=ax)

ax.set_title('User distribution across age and gender')
ax.set_xlabel('Number of users')

density_plot = (
    data_gender_age.groupby(['sex', 'age'])
    .size()
    .reset_index(name='count')
)

fig, ax = plt.subplots(figsize=(8, 6))

sns.lineplot(data=density_plot, x='age', y='count', hue='sex', ci=None)

plt.title('Density distribution of age across both genders')
plt.xlabel('Age')
plt.ylabel('Count')

plt.tight_layout()

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming your DataFrame is named 'cupid_df'
columns_of_interest = ['age', 'status', 'sex', 'orientation', 'body_type', 'diet', 'drinks', 'drugs', 'education',
                       'height', 'income', 'smokes']

# Subset the DataFrame to the columns of interest
subset_data = cupid_df[columns_of_interest]

# Encode categorical columns
encoder = LabelEncoder()
categorical_columns = ['status', 'sex', 'orientation', 'body_type', 'diet', 'drinks', 'drugs', 'education', 'smokes']
subset_data[categorical_columns] = subset_data[categorical_columns].apply(lambda x: encoder.fit_transform(x.astype(str)))

# Calculate the correlation matrix
correlation_matrix = subset_data.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
