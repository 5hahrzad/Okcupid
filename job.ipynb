```python
## importing libraries

import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import plotly.graph_objects as go
%matplotlib inline
```

```python
## Loading csv into pandas dataframe
cupid_df = pd.read_csv(r"C:\Users\sghad\OneDrive\Desktop\DA\Final Project\data_set\okcupid_data.csv")
```

```python
missing_values_job = cupid_df['job'].isnull().sum()
print("Number of missing values in the 'job' column:", missing_values_job)
```

```python
# Replace similar categories with a common format
cupid_df['job'] = cupid_df['job'].str.replace('manager', 'management')
cupid_df['job'] = cupid_df['job'].str.replace('executive', 'management')
cupid_df['job'] = cupid_df['job'].str.replace('supervisor', 'management')

# Replace remaining variations with a standard category
cupid_df['job'] = cupid_df['job'].replace(['other', 'student'], 'unemployed')

# Check the cleaned job categories
print(cupid_df['job'].unique())
```

```python
cupid_df.dropna(subset=['job'], inplace=True)
```

```python
most_frequent_job = cupid_df['job'].mode()[0]
cupid_df['job'].fillna(most_frequent_job, inplace=True)
```

```python
missing_values_job = cupid_df['job'].isnull().sum()
missing_values_job
```

```python
# Calculate the value counts of job categories
job_counts = cupid_df['job'].value_counts()

# Plot the distribution of individuals by job category
plt.figure(figsize=(12, 6))
job_counts.plot(kind='bar')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.title('Distribution of Individuals by Job Category')
plt.xticks(rotation=45)
plt.show()
```
